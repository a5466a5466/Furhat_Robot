## Data Preprocessing

```python
import pandas as pd
data = pd.read_csv("dataset.csv")

labels = data["emotion"]
inputs = data.drop("emotion", axis=1)
```

### Split Data 
â€¢	Train/val/test : 70/20/10

```python
from sklearn.model_selection import train_test_split


# split = 70/20/10
data_in, test_in, data_out, test_out = train_test_split(
    inputs,
    labels,
    test_size=0.1,
    random_state=42,
    stratify=labels
)

train_in, val_in, train_out, val_out = train_test_split(
    data_in,
    data_out,
    test_size=(0.2/0.9),  # 20% of the original data
    random_state=42,
    stratify=data_out
)

print("\nLenght of each split of the data: ", len(train_in), len(val_in), len(test_in), "\n")
```

### Normalization

```python
from sklearn import preprocessing

normalized_train_in = preprocessing.scale(train_in)
normalized_val_in = preprocessing.scale(val_in)
```

## Chosing Model

```python
from sklearn.metrics import accuracy_score
```

```python
def train_model(model, train_in = train_in, train_out = train_out):
    model.fit(train_in, train_out)

def eval_model(model, val_in = val_in, val_out=val_out):
    predicted_val = model.predict(val_in)
    return accuracy_score(predicted_val, val_out)
```

### SGDClassifier

```python
from sklearn.linear_model import SGDClassifier

SGDClassifier_Model = SGDClassifier()

train_model(SGDClassifier_Model)
print("SGDClassifier_Model accuracy_score: ", eval_model(SGDClassifier_Model))
```

### SVC

```python
from sklearn.svm import SVC

SVC_model = SVC()
train_model(SVC_model)
print("SVC accuracy_score: ", eval_model(SVC_model))
```

### KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
train_model(knn)
print("KNeighborsClassifier accuracy_score: ", eval_model(knn))
```

## Try Normalized Data

```python
SGDClassifier_Model = SGDClassifier()

train_model(SGDClassifier_Model, normalized_train_in)
print("SGDClassifier_Model accuracy_score: ", eval_model(SGDClassifier_Model, normalized_val_in))
```

### N_SVC

```python
N_SVC = SVC()
train_model(N_SVC, train_in=normalized_train_in)
eval_model(N_SVC, val_in=normalized_val_in)
```

### N_KNeighborsClassifier

```python
N_knn = KNeighborsClassifier()
train_model(N_knn, train_in=normalized_train_in)
eval_model(N_knn, val_in=normalized_val_in)
```

### N GaussianNB

```python
from sklearn.naive_bayes import GaussianNB

N_GaussianNB = GaussianNB()
train_model(N_GaussianNB, train_in=normalized_train_in)
eval_model(N_GaussianNB, val_in=normalized_val_in)
```

# Hyperparameter searching/tuning

```python
from sklearn.model_selection import GridSearchCV
```

### SGDClassifier tuning

```python
SGD_param_grid = [
    {"loss": ['hinge', 'log_loss', 'loss', 'squared_error'], "penalty" : ['l2', 'l1']}
]
best_SGD = GridSearchCV(SGDClassifier(), SGD_param_grid)
train_model(best_SGD, normalized_train_in)
eval_model(best_SGD, normalized_val_in)
```

```python
print(
    "Best parameters of SVC model: ",
    best_SGD.best_params_
)
```

### Tuing KNeighborsClassifier

```python
Knn_param_grid = [
    {"algorithm": ['auto', 'ball_tree', 'kd_tree', 'brute'], "n_neighbors": [1, 5, 10, 25, 50]}
]

best_knn = GridSearchCV(KNeighborsClassifier(), Knn_param_grid)
best_knn.fit(normalized_train_in, train_out)
eval_model(best_knn, normalized_val_in)
```

```python
print(
    "Best parameters of SVC model: ",
    best_knn.best_params_
)
```

### Tuning SVC

```python
SVC_param_grid = [
    {"kernel": ["rbf", "linear", "sigmoid"], "degree": [3, 10, 25, 50], "gamma":["scale", "auto"]}
]

best_SVC = GridSearchCV(SVC(), SVC_param_grid)
best_SVC.fit(normalized_train_in, train_out)
eval_model(best_SVC, normalized_val_in)
```

```python
print("Best parameters of SVC model: ",
    best_SVC.best_params_)
```
